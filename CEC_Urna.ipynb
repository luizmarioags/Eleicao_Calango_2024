## Lista Geral

```python
#Testando split de votos
def separar_votos(df):
    # Inicializar as listas para cada coluna de votos
    voto_lista1_geral = []
    voto_lista2_geral = []
    voto_lista3_geral = []

    # Iterar sobre os votos na coluna do DataFrame
    for voto in df['Quais são os seus votos para a Lista Geral ?\nVocê tem 3 votos!']:
        # Dividir os votos com base na vírgula
        votos_separados = voto.split(", ")

        # Adicionar cada música à coluna correspondente
        for i, musica in enumerate(votos_separados):
            if i == 0:
                voto_lista1_geral.append(musica)
            elif i == 1:
                voto_lista2_geral.append(musica)
            elif i == 2:
                voto_lista3_geral.append(musica)

    return voto_lista1_geral, voto_lista2_geral, voto_lista3_geral
# Chamar a função para separar os votos
voto_lista1_geral, voto_lista2_geral, voto_lista3_geral = separar_votos(df)
df['Voto 1 - Lista Geral'] = voto_lista1_geral
df['Voto 2 - Lista Geral'] = voto_lista2_geral
df['Voto 3 - Lista Geral'] = voto_lista3_geral
```

```python
df
```

```python
# Função para agregar votos por música
def contar_votos(voto_lista1_geral, voto_lista2_geral, voto_lista3_geral):
    # Criar um dicionário para armazenar os votos por música
    votos_por_musica = {}

    # Iterar sobre as listas de votos e contar os votos para cada música
    for lista in [voto_lista1_geral, voto_lista2_geral, voto_lista3_geral]:
        for musica in lista:
            if musica in votos_por_musica:
                votos_por_musica[musica] += 1
            else:
                votos_por_musica[musica] = 1

    return votos_por_musica
# Agregar os votos por música
votos_por_musica = contar_votos(voto_lista1_geral, voto_lista2_geral, voto_lista3_geral)

# Criar um DataFrame a partir do dicionário de votos por música
df_votos_agregados = pd.DataFrame(list(votos_por_musica.items()), columns=['Música', 'Total de Votos'])

# Imprimir o DataFrame de votos agregados
print(df_votos_agregados)
```

```python
sns.histplot(x = 'Total de Votos',data = df_votos_agregados)
```

```python
#Critério da média
media = np.mean(df_votos_agregados['Total de Votos']) #Cálculo da Média
df_votos_agregados['Critério Média'] = np.where(df_votos_agregados['Total de Votos'] <=
                                media, 'Reprovada', 'Aprovada')
#Percentil
#Percentil
percentil_75 = np.quantile(df_votos_agregados['Total de Votos'],0.75)
df_votos_agregados['Percentil'] = np.where(df_votos_agregados['Total de Votos'] > percentil_75, True, False)
```

```python
df_votos_agregados
```

```python
df_votos_agregados.describe()
```

```python
#Filtrando para mostrar os vencedores
df_1_turno = df_votos_agregados[
    (df_votos_agregados['Critério Média'] == 'Aprovada') & (df_votos_agregados['Percentil'] == True)]
df_1_turno.sort_values(by = 'Total de Votos',ascending = False)
```

```python
df_1_turno.sort_values(by = 'Total de Votos',ascending = False)
```

```python
musicas_1_turno = list(df_1_turno['Música'])
```

```python
musicas_1_turno
```

```python
# Filtrar o DataFrame original para incluir apenas as linhas com as músicas da lista
df_plot_musicas = df[df['Voto 1 - Lista Geral'].isin(musicas_1_turno) |
                  df['Voto 2 - Lista Geral'].isin(musicas_1_turno) |
                  df['Voto 3 - Lista Geral'].isin(musicas_1_turno)]
```

```python
df_plot_musicas
```

```python
df_plot_musicas.columns
```

```python
# Definir marcadores para cada naipe
naipes = df_plot_musicas['Qual seu Naipe ?'].unique()
markers = ['o', 's', 'D', '^']
default_marker = 'o'  # Marcador padrão para um naipe não reconhecido
naipe_markers = {naipe: marker for naipe, marker in zip(naipes, markers)}

# Para cada música do primeiro turno
for musica in musicas_1_turno:
    # Filtrar o DataFrame para incluir apenas as linhas com a música atual
    df_musica = df_plot_musicas[df_plot_musicas[['Voto 1 - Lista Geral', 'Voto 2 - Lista Geral', 'Voto 3 - Lista Geral']].isin([musica]).any(axis=1)]

    # Converter a coluna 'Carimbo de data/hora' para o tipo datetime
    df_musica['Carimbo de data/hora'] = pd.to_datetime(df_musica['Carimbo de data/hora'])

    # Agrupar as músicas por data/hora e naipe e contar o número de vezes que aparecem
    df_votos_por_dia_naipe = df_musica.groupby([df_musica['Carimbo de data/hora'].dt.strftime('%d/%m/%Y'), 'Qual seu Naipe ?']).size().unstack(fill_value=0)

    # Calcular os totais acumulados de votos por data/hora e naipe
    df_totais_acumulados_naipe = df_votos_por_dia_naipe.cumsum()

    # Plotar os totais acumulados de votos por Carimbo de data/hora e naipe
    plt.figure(figsize=(12, 8))
    for naipe in df_totais_acumulados_naipe.columns:
        marker = naipe_markers.get(naipe, default_marker)
        plt.plot(df_totais_acumulados_naipe.index, df_totais_acumulados_naipe[naipe], label=naipe, marker=marker)

    plt.title(f'Quantidade acumulada de votos por dia para a música "{musica}"')
    plt.xlabel('Dia/Mês')
    plt.ylabel('Quantidade de votos acumulada')
    plt.xticks(rotation=45)
    plt.legend(title='Naipe', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
# Filtrar o DataFrame para incluir apenas as linhas com as músicas do primeiro turno
df_filtrado = df_plot_musicas

# Converter a coluna 'Carimbo de data/hora' para o tipo datetime e extrair apenas o dia/mês/ano
df_filtrado['Carimbo de data/hora'] = pd.to_datetime(df_filtrado['Carimbo de data/hora']).dt.strftime('%d/%m/%Y')

# Inicializar um DataFrame para armazenar os totais acumulados de votos por data/hora para cada música
df_totais_acumulados = pd.DataFrame(index=df_filtrado['Carimbo de data/hora'].unique(), columns=musicas_1_turno).fillna(0)

# Calcular os totais acumulados de votos por data/hora para cada música
for musica in musicas_1_turno:
    df_musica = df_filtrado[(df_filtrado['Voto 1 - Lista Geral'] == musica) |
                            (df_filtrado['Voto 2 - Lista Geral'] == musica) |
                            (df_filtrado['Voto 3 - Lista Geral'] == musica)]
    df_votos_por_dia = df_musica.groupby('Carimbo de data/hora').size()
    df_totais_acumulados[musica] = df_votos_por_dia.cumsum()

# Preencher valores ausentes com a quantidade acumulada do dia anterior apenas se o valor do dia atual for zero
df_totais_acumulados = df_totais_acumulados.fillna(method='ffill')

# Definir uma lista de marcadores diferentes para cada música
marcadores = ['o', 's', 'D', '^', 'v', 'p', '*', 'X', 'h', 'd']

# Plotar os totais acumulados de votos por Carimbo de data/hora para cada música
plt.figure(figsize=(12, 8))
for i, musica in enumerate(musicas_1_turno):
    plt.plot(df_totais_acumulados.index, df_totais_acumulados[musica], label=musica, marker=marcadores[i])

plt.title('Quantidade acumulada de votos por música ao longo do tempo')
plt.xlabel('Data')
plt.ylabel('Quantidade de votos acumulada')
plt.xticks(rotation=45)
plt.legend(title='Música',bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
